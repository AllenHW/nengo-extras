```python
import matplotlib.pyplot as plt
%matplotlib inline
import nengo
import numpy as np

import nengo_extras.matplotlib as neplt
from nengo_extras.data import load_cifar10, one_hot_from_labels
from nengo_extras.vision import Gabor, Mask

shape = (3, 32, 32)


def imgs_for_show(imgs):
    imgs = np.transpose(imgs.reshape(-1, 3, 32, 32), (0, 2, 3, 1))
    return imgs[0] if imgs.shape[0] == 1 else imgs


def imshow(img, vmin=-1, vmax=1, **kwargs):
    neplt.imshow(imgs_for_show(img), vmin=vmin, vmax=vmax, **kwargs)


rng = np.random.RandomState(9)
```

```python
# --- load the data
(X_train, y_train), (X_test, y_test) = load_cifar10()

# normalize to -1 to 1
X_train = (2./256)*X_train - 1
X_test = (2./256)*X_test - 1

# create one-hot targets
train_targets = one_hot_from_labels(y_train, classes=10)
test_targets = one_hot_from_labels(y_test, classes=10)
```

```python
X_rectest = X_test[:10]
```

## Create the network

We choose not to randomize the intercepts and max rates, so that our results are only affected by the different encoder choices.

We create a function ``get_outs`` that returns the output of the network, so that we can evaluate the network statically on many images (i.e. we don't need to run the simulator).

```python
# --- set up network parameters
n_vis = X_train.shape[1]
n_out = train_targets.shape[1]
#n_hid = 100
# n_hid = 300
n_hid = 1000
# n_hid = 3000

ens_params = dict(
    eval_points=X_train,
    neuron_type=nengo.LIFRate(),
    intercepts=nengo.dists.Choice([-0.5]),
    max_rates=nengo.dists.Choice([100]),
    )

solver = nengo.solvers.LstsqL2(reg=0.01)
# solver = nengo.solvers.LstsqL2(reg=0.0001)

with nengo.Network(seed=3) as model:
    a = nengo.Ensemble(n_hid, n_vis, **ens_params)
    v = nengo.Node(size_in=n_out)
    conn = nengo.Connection(
        a, v, synapse=None,
        eval_points=X_train, function=train_targets, solver=solver)
    
    w = nengo.Node(size_in=n_vis)
    conn_rec = nengo.Connection(
        a, w, synapse=None,
        eval_points=X_train, function=X_train, solver=solver)
    
def get_outs(sim, images):
    _, acts = nengo.utils.ensemble.tuning_curves(a, sim, inputs=images)
    return np.dot(acts, sim.data[conn].weights.T), np.dot(acts, sim.data[conn_rec].weights.T)

def get_error(sim, images, labels):
    outs, rec_outs = get_outs(sim, images)
    error = np.argmax(outs, axis=1) != labels
    rec_error = nengo.utils.numpy.rms(rec_outs - images, axis=1)
    return error, rec_error
```

## Normally distributed encoders

These are the standard encoders used in the NEF. Since our data is high-dimensional, they have a lot of space to cover, and do not work particularly well.

```python
encoders = rng.normal(size=(n_hid,) + shape).reshape(n_hid, -1)
a.encoders = encoders

neplt.tile(imgs_for_show(encoders[:20]), rows=4, cols=5, vmin=encoders.min(), vmax=encoders.max(), grid=True)
```

```python
with nengo.Simulator(model) as sim:
    test_error, test_rec_error = get_error(sim, X_test, y_test)
    print("Test error: %0.2f%%, %0.2f" % (100*test_error.mean(), test_rec_error.mean()))
```

```python
_, recs = get_outs(sim, X_rectest)
neplt.compare((imgs_for_show(X_rectest), imgs_for_show(recs)), vmin=-1, vmax=1)
```

## Normally distributed sparse encoders

The same as before, but now each encoder has a limited receptive field. This makes each neuron only responsible for part of the image, and allows them to work together better to encode the whole image.

```python
encoders = rng.normal(size=(n_hid, shape[0], 11, 11))
encoders = Mask(shape).populate(encoders, rng=rng, flatten=True)
a.encoders = encoders

neplt.tile(imgs_for_show(encoders[:20]), rows=4, cols=5, vmin=encoders.min(), vmax=encoders.max(), grid=True)
```

```python
with nengo.Simulator(model) as sim:
    test_error, test_rec_error = get_error(sim, X_test, y_test)
    print("Test error: %0.2f%%, %0.2f" % (100*test_error.mean(), test_rec_error.mean()))
```

```python
_, recs = get_outs(sim, X_rectest)
neplt.compare((imgs_for_show(X_rectest), imgs_for_show(recs)), vmin=-1, vmax=1)
```

## Gabor filter encoders

Neurons in primary visual cortex have tuning that resembles Gabor filters. We use Gabor filters that cover patches of the image. This way, each neuron is only responsible for part of the image, and the statistics of the Gabor filters roughly match the image statistics.

```python
gabors = Gabor().generate(n_hid, (13, 13), rng=rng)
colors = nengo.dists.UniformHypersphere(surface=True).sample(n_hid, shape[0], rng=rng)
color_gabors = gabors[:, None, :, :] * colors[:, :, None, None]
encoders = Mask(shape).populate(color_gabors, rng=rng, flatten=True)
a.encoders = encoders

neplt.tile(imgs_for_show(encoders[:20]), rows=4, cols=5, vmin=encoders.min(), vmax=encoders.max(), grid=True)
```

```python
with nengo.Simulator(model) as sim:
    test_error, test_rec_error = get_error(sim, X_test, y_test)
    print("Test error: %0.2f%%, %0.2f" % (100*test_error.mean(), test_rec_error.mean()))
```

```python
_, recs = get_outs(sim, X_rectest)
neplt.compare((imgs_for_show(X_rectest), imgs_for_show(recs)), vmin=-1, vmax=1)
```
